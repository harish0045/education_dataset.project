```python
continent_quality = df1.groupby("Continent")["Education_Quality"].mean()
plt.figure(figsize=(8, 8))
continent_quality.plot.pie(
    autopct='%1.2f%%',           # Show percentage
    startangle=90,               # Rotate start
    shadow=True,                 # Add shadow effect
    explode=[0.05]*len(continent_quality),  # Separate slices slightly
    labeldistance=1.1)           # Adjust label distance
plt.title("Average Education Quality Continent wise")
plt.ylabel("")  
plt.tight_layout()
plt.show()
```

```python
continent_quality = top_10.groupby("Continent")["Education_Quality"].mean()
plt.figure(figsize=(8, 8))
continent_quality.plot.pie(
    autopct='%1.2f%%',           # Show percentage
    startangle=90,               # Rotate start
    shadow=True,                 # Add shadow effect
    explode=[0.05]*len(continent_quality),  # Separate slices slightly
    labeldistance=1.1)           # Adjust label distance
plt.title("Average Education Quality Continent by top 10 countries")
plt.ylabel("")  
plt.tight_layout()
plt.show()
```

```python
continentquality = top_10.groupby("Education_System")["Education_Quality"].mean()
plt.figure(figsize=(8, 8))
continentquality.plot.pie(
    autopct='%1.2f%%',           # Show percentage
    startangle=90,               # Rotate start
    shadow=True,                 # Add shadow effect
    labeldistance=1.1)            # Adjust label distance
plt.title("education methods used by top 10 countries")
plt.ylabel("")  
plt.tight_layout()
plt.show()
```
